## Challenge

To explore tokenization beyond the basic examples, I experimented with a sentence containing different types of text, including emojis, URLs, email addresses, numbers, and punctuation.
The goal of this challenge was to observe how NLTK's `word_tokenize()` function handles different types of tokens in real-world text.


**Key Learning:**  
Experimenting with diverse text inputs provides a better understanding of tokenization and prepares us for handling real-world NLP datasets.

```python
corpus =""" Hi 😊! welcom to https://github.com. Email me at muskannarainmaurya@gmail.com. I i have a cgpa above 8.5!"""
```

```python
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize,wordpunct_tokenize
```

```python
document=sent_tokenize(corpus)
print(document)
```

```python
for sentences in document:
    print(word_tokenize(sentences))
```

```python
for sentences in document:
    print(wordpunct_tokenize(sentences))
```

Observation

- Emojis are treated as separate tokens.
- Punctuation marks are tokenized separately after using "wordpunct_tokenize".
- Numbers and symbols are recognized as individual tokens.
- This experiment helped me understand how tokenization works on text that resembles real-world data rather than simple sentences.
