```python
text =['Students','studying','happily','running','towards','university','adjustable','chairs','arranged','beautifully']

```

```python
from nltk.stem import PorterStemmer,RegexpStemmer,SnowballStemmer
```

```python
stemming=PorterStemmer()
reg_stemmer=RegexpStemmer("ing$|s$|able$",min=4)
ss_stemmer=SnowballStemmer('english')
```

```python
for words in text:
    print(words+'--->'+stemming.stem(words))
```

```python
for words in text:
    print(words+'--->'+reg_stemmer.stem(words))
```

```python
for words in text:
    print(words+'--->'+ss_stemmer.stem(words))
```
