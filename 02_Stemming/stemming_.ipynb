###stemming
getting to the root word
- plays--------> play
- playing------> play

### Porter stemmer


```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
words=["playing","eats","crossing","eaten","writing","history","programming"]
```

```python
for word in words:
    print(word+"---->"+stemming.stem(word))
```

### Regexpstemmer
- ("ing$|s$|able$",min=4)

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer("ing$|s$|able$", min=4)

for word in words:
    print(word+'---->'+ reg_stemmer.stem(word))
```

### SnowballStemmer

```python
from nltk.stem import SnowballStemmer
SB_stemmer = SnowballStemmer('english')

for word in words:
    print(word+'---->'+SB_stemmer.stem(word))
```

```python
SB_stemmer.stem("lovingly"),SB_stemmer.stem("fairly"),SB_stemmer.stem("sincerely")
```
